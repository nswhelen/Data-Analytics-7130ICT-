### Calculate page rank by using simple degree centrality

```python
G = nx.karate_club_graph()

def degree_centrality(G):
    p = np.array([G.degree(index) for node, index in enumerate(G.nodes())])
    return p.tolist()

centralities = degree_centrality(G)

fig, ax = plt.subplots()
nx.draw(G, cmap=plt.get_cmap('coolwarm'), node_color=centralities, with_labels=True, ax=ax)
plt.show()
```

### Calculate page rank by updating the centrality of each node with the iteration

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

def pagerank_centrality(G, iter=100):
    p = np.array([1 for i in list(G.nodes())])
    for k in range(iter):
        for i in G.nodes():
            for j in G.nodes():
                # update the centrality
                try:
                    p[i] += int(G.number_of_edges(i, j) * p[j] / G.degree[j])
                except:
                    pass
        
        norm = sum(p)
        p = p / norm        
    return list(p)

G = nx.karate_club_graph()
centralities = pagerank_centrality(G)
plt.show()

nx.draw(G, cmap=plt.get_cmap('coolwarm'), node_color=centralities, with_labels=True, ax=ax)

centralities
```
